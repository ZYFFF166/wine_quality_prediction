```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data/winequality-white.csv', sep =";")
df.head(5)
```

```python
def define_labels(input):
    if(input <=5):
        return 0
    elif(input >5 and input <7):
        return 1
    else:
        return 2
labels = df.quality.apply(define_labels)
df = df.drop(['quality'], axis =1)

```

```python
input_data = df[['volatile acidity', 'density', 'pH', 'sulphates', 'alcohol', 'chlorides']]

# mine
# input_data = df[["volatile acidity", "residual sugar", "free sulfur dioxide", "sulphates", "alcohol"]]

```

```python
X_train, X_test, y_train, y_test = train_test_split(input_data, labels, test_size =0.2, random_state= 118) #118
```

```python
class LogisticRegression1:
    def __init__(self, lr=0.01, num_iter=10000, fit_intercept=True, verbose=True, class_num = 2):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
        self.class_num = class_num
        self.loss_data= []
    
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def __loss(self, h, y):
        #h = np.squeeze(np.asarray(h))
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        self.thetas = []
        ys = []
        class_num = self.class_num
        
        if class_num == 2 :
            self.thetas = [np.zeros(X.shape[1])]
            ys = [y]
        else:
            for i in range(class_num):
                self.thetas.append(np.zeros(X.shape[1]))
                result = pd.core.series.Series(map(lambda y: 1 if y == i else 0, y.copy()))
                ys.append(result)
        
        model_needed = 1 if class_num == 2 else class_num
        for j in range(model_needed):
            self.loss_data.append([])
            self.loss_data[j].append([])
            self.loss_data[j].append([])
            for i in range(self.num_iter):
                z = np.dot(X, self.thetas[j])
                h = self.__sigmoid(z)
                #h = np.squeeze(np.asarray(h))
                gradient = np.dot(X.T, (h - ys[j])) / ys[j].size
                gradient = np.squeeze(np.asarray(gradient))
                self.thetas[j] -= self.lr * gradient

                z = np.dot(X, self.thetas[j])
                h = self.__sigmoid(z)
                loss = self.__loss(h, ys[j])


                if(self.verbose ==True and i % 1000 == 0):
                    self.loss_data[j][0].append(i)
                    self.loss_data[j][1].append(loss)
                    print(f'times:{i}, loss: {round(loss, 4)} \t')
    
    def predict_prob(self, X, model_index = 0):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.__sigmoid(np.dot(X, self.thetas[model_index]))
    
    def draw_loss(self):
        model.loss_data
        fig=plt.figure(figsize=(9, 8), dpi= 80, facecolor='w', edgecolor='k')

        plt.plot(model.loss_data[0][0], model.loss_data[0][1])
        plt.plot(model.loss_data[1][0], model.loss_data[1][1])
        plt.plot(model.loss_data[2][0], model.loss_data[2][1])

        plt.ylabel("Loss")
        plt.xlabel("Num of iterations")
        plt.legend(('Model for Low', 'Model for Medium', 'Model for High'))
        plt.title("Loss change of 3 model with the increase of iterations")
        plt.savefig('./books_read.png')
        plt.show()

    
    def predict(self, X):
        return self.predict_prob(X).round()
    
    def predict(self, X):
        model_num = 1 if self.class_num == 2 else self.class_num
        probs = []
        for i in range(model_num):
            probs.append(self.predict_prob(X, i))
            
        result = []
        for i in range(len(probs[0])):
            cur_max = -1
            cur_max_idx = -1
            
            for j in range(self.class_num):
                if(probs[j][i] > cur_max):
                    cur_max = probs[j][i]
                    cur_max_idx = j
            
            result.append(cur_max_idx)
        return pd.Series(result)
```

```python
model = LogisticRegression1(class_num=3, num_iter=50000)
model.fit(X_train, y_train)
predictions_LR = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(predictions_LR, y_test) )
# predictions_LR.hist()
```
